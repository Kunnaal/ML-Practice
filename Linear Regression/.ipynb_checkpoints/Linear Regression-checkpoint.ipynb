```python
import pandas
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv("./cost_revenue_clean.csv")
```

```python
data
```

```python
data.describe()
```

```python
print(5.034e3) #Scientific notation
```

```python
#Production Budgets
X = pandas.DataFrame(data ,columns=['production_budget_usd'])

#Revenue
y = pandas.DataFrame(data, columns=['worldwide_gross_usd'])
```

Visualizing

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film cost v/s Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,)
plt.xlim(0,)
plt.show()
```

## Implementation

```python
regression = LinearRegression()
```

```python
%%capture
regression.fit(X, y) #linear regression, calculating params.
```

```python
regression.coef_ #Slope coefficient of regression
```

```python
regression.intercept_ #intercept of regression
#This intercept means that a movie with a budget of 0 will loose 7 M$!!!
```

```python
#Adding line to graph

plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

plt.plot(X, regression.predict(X), color='red', linewidth=2)

plt.title('Film cost v/s Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,)
plt.xlim(0,)
plt.show()
```

```python
#Conclusion: Predictions of out model works very worse at extreme values.
#It is an oversimplification of real life model.
```

### Goodness of fit / R^2

```python
print(str(regression.score(X, y)*100)+ '%')
```

```python
# 55% score is probably good for such a simple model
```
