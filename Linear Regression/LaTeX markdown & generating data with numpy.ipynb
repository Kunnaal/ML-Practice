### Notebook imports & Magic functions

**Magic functions**: IPython has a set of predefined ‘magic functions’ that you can call with a command line style syntax. There are two kinds of magics, line-oriented and cell-oriented. Line magics are prefixed with the % character and work much like OS command-line calls: they get as an argument the rest of the line, where arguments are passed without parentheses or quotes. Lines magics can return results and can be used in the right hand side of an assignment. Cell magics are prefixed with a double %%, and they are functions that get as an argument not only the rest of the line, but also the lines below it in a separate argument.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that 
# produced it. The resulting plots will then also be stored in the notebook document.
```

### Simple cost function:
$y = x^2 + x + 1$

```python
#Function definition:
def f(x):
    return x**2 + x + 1
```

```python
#Generating data:
x_1 = np.linspace(start = -3, stop = 3, num = 500)
```

```python
#Plotting cost function on curve:
plt.title('Cost function')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.plot(x_1, f(x_1))
plt.show()
```

### Slope & Derivatives

```python
#df(x), function to calculate the derivative of f(x):
def df(x):
    return 2*x + 1
```

```python
#Plotting cost function & derivative of cost function side by side:

plt.figure(figsize = [15, 5]) #Setting height & width of plot to 3x3 inches.

#figure 1 (cost function):
plt.subplot(1, 2, 1) #(row, column, index)
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('Cost function')
plt.plot(x_1, f(x_1), color = 'red')

#figure 2 (derivative of cost function):
plt.subplot(1, 2, 2) #(row, column, index)
plt.xlabel('X')
plt.ylabel('df(x)')
plt.title('Derivative of Cost function')
plt.plot(x_1, df(x_1), color = 'blue')
plt.grid()

#Show both the plots:
plt.show()
```

### Gradient descent

```python
#Gradient Descent
new_x = 3
prev_x = 0
alpha = 0.1 #Step multiplier
precision = 0.00001

#Starting values of lists:
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - gradient*alpha
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - prev_x)
    
    if(step_size < precision):
        print('No. of times ran:', n)
        break

print('The local minimum reached:', new_x)
print('The slope (differential) at local minimum:', df(new_x))
print('The Cost at local minimum:', f(new_x))
```

```python
#Superimposing gradient descent on cost & derivative function for visualization:

plt.figure(figsize = [15, 10]) #Setting height & width of plot to 3x3 inches.

#figure 1 (cost function):
plt.subplot(2, 2, 1) #(row, column, index)
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('Cost function')
plt.plot(x_1, f(x_1), color = 'red')
plt.scatter(x_list, f(np.array(x_list)), marker = 'x', alpha = 0.6) #np.array as our function f() cannot work with list.

#figure 2 (derivative of cost function):
plt.subplot(2, 2, 2) #(row, column, index)
plt.xlabel('X')
plt.ylabel('df(x)')
plt.title('Derivative of Cost function')
plt.plot(x_1, df(x_1), color = 'blue')
plt.scatter(x_list, slope_list, color='red', marker = '*', alpha = 0.6)
plt.grid()

#figure 3 (Close up of derivative):
plt.subplot(2, 2, 3) #(row, column, index)
plt.xlabel('X')
plt.ylabel('df(x)')
plt.title('Close up of derivative function')
plt.xlim(-0.55, -0.20)
plt.ylim(-0.1, 0.6)
plt.plot(x_1, df(x_1), color = 'blue')
plt.scatter(x_list, slope_list, color='red', marker = '*', alpha = 0.6)
plt.grid()

#Show both the plots:
plt.show()
```
